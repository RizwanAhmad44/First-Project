```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.DataFrame({
    'Rel Hum_%': [55, 60, 45, 70, 52],
    'Visibility_km': [35, 42, 38, 45, 50],
    'Weather Condition': ['Clear', 'Clear', 'Cloudy', 'Clear', 'Rainy']
}).assign(**{'Date': pd.date_range(start='2023-01-01', periods=5)}).set_index('Date')

filtered_data = data[(data['Weather Condition'] == 'Clear') & 
                    ((data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40))]
plt.figure(figsize=(10, 6))
sc = plt.scatter(filtered_data['Rel Hum_%'], filtered_data['Visibility_km'],
                 c=filtered_data.index.hour,  # Works because we created datetime index
                 cmap='viridis',
                 alpha=0.8)
plt.colorbar(sc, label='Hour of Day')
plt.title('Filtered: Clear Weather & (Humidity >50% or Visibility >40km)')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Visibility (km)')
plt.grid(True)
plt.show()
```
